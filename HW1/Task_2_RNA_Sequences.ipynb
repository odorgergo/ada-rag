We start by reading in the metadata and also reading the data into a list of dataframes. We make sure that the dataframes are indexed by the names of the speices. 

```python
import pandas as pd

meta = pd.read_excel("Data/microbiome/metadata.xls")

list_ = []
for i in range(9):
    temp = pd.read_excel("Data/microbiome/%s.xls" % meta.BARCODE[i],header=None)
    temp.columns=['names','value%d' %i]
    list_.append(temp.set_index(['names']))
    print(list_[i].index.is_unique)
```

Now we outer merge all the dataframe in the list by the index.

```python
data = list_[0].copy()
for i in range(1,9):
    data=data.merge(list_[i],left_index=True, right_index=True,how='outer')
```

Finally, we replace NaN values by the tag unknown and we rename the columns which automatically have hiearchical indexing.

```python
data.fillna('unknown', inplace=True)
data.columns=[meta.GROUP,meta.SAMPLE]
```

We can see that all indices are unique and also that the columns are named correctly.

```python
print("Indices are unique: " + str(data.index.is_unique))
data
```
